# ****Data Cleaning****

```python
# Load the clean data from the Lakehouse table into a Spark DataFrame
df_clean = spark.read.table("CustomerChurnAnalytics.DF_Clean_Churn_Data")

# Display a few rows to make sure it loaded correctly
display(df_clean)
```

```python
# Select all the columns that describe the customer
dim_customer_df = df_clean.select(
    "customerID", "gender", "SeniorCitizen", "Partner", "Dependents",
    "PhoneService", "MultipleLines", "InternetService", "OnlineSecurity",
    "OnlineBackup", "DeviceProtection", "TechSupport", "StreamingTV",
    "StreamingMovies", "Contract", "PaperlessBilling", "PaymentMethod"
)

# Save this DataFrame as the 'dim_customer' table
dim_customer_df.write.mode("overwrite").format("delta").save("Tables/dim_customer")
print("Successfully created the 'dim_customer' table.")
```

```python
# Select the keys and the numeric measures for our fact table
fact_churn_df = df_clean.select(
    "customerID", "Churn", "tenure", "MonthlyCharges", "TotalCharges"
)

# Save this DataFrame as the 'fact_churn' table
fact_churn_df.write.mode("overwrite").format("delta").save("Tables/fact_churn")
print("Successfully created the 'fact_churn' table.")
```
